```python
import time
import random

# Function to detect eye closure duration
def detect_eye_closure_duration():
    # Simulate detection of eye closure duration
    return random.uniform(0, 2)  # Random value between 0 and 2 seconds

# Function to detect head movement
def detect_head_movement():
    # Simulate detection of head movement
    return random.uniform(0, 30)  # Random value between 0 and 30 degrees

# Function to detect yawn frequency
def detect_yawn_frequency():
    # Simulate detection of yawn frequency
    return random.randint(0, 5)  # Random integer between 0 and 5 yawns per minute

# Function to detect steering wheel deviation
def detect_steering_wheel_deviation():
    # Simulate detection of steering wheel deviation
    return random.uniform(0, 20)  # Random value between 0 and 20 degrees

# Constants
DROWSY_THRESHOLD = 0.5  # Threshold for determining drowsiness (you may adjust this value)
NUM_SAMPLES = 50  # Number of samples to generate

# Generate and print data samples
print("Generating Data Samples:\n")
print("{:<25} {:<20} {:<25} {:<30} {:<15}".format("Eye Closure Duration", "Head Movement", "Yawn Frequency", "Steering Wheel Deviation", "Label"))

for i in range(NUM_SAMPLES):
    # Detect eye closure duration
    eye_closure_duration = detect_eye_closure_duration()

    # Detect head movement
    head_movement = detect_head_movement()

    # Detect yawn frequency
    yawn_frequency = detect_yawn_frequency()

    # Detect steering wheel deviation
    steering_deviation = detect_steering_wheel_deviation()

    # Determine if the driver is drowsy or alert based on the collected data
    if (eye_closure_duration > DROWSY_THRESHOLD or
        head_movement > DROWSY_THRESHOLD or
        yawn_frequency > DROWSY_THRESHOLD or
        steering_deviation > DROWSY_THRESHOLD):
        label = "Drowsy"
    else:
        label = "Alert"

    # Print the collected data and label
    print("{:<25} {:<20} {:<25} {:<30} {:<15}".format(round(eye_closure_duration, 2), round(head_movement, 2), yawn_frequency, round(steering_deviation, 2), label))

    # Add a small delay to simulate real-time operation
    time.sleep(0.1)

print("\nData Generation Complete.")
```
